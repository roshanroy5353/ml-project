```python
df['Weather'].fillna(df['Weather'].mode()[0], inplace=True)
```

```python
df['Time_of_Day'].fillna(df['Time_of_Day'].mode()[0], inplace=True)
```

```python
df['Traffic_Density'].fillna(df['Traffic_Density'].mode()[0], inplace=True)
```

```python
df['Road_Type'].fillna(df['Road_Type'].mode()[0], inplace=True)
```

```python
df['Road_Condition'].fillna(df['Road_Condition'].mode()[0], inplace=True)
```

```python
df['Accident_History'].fillna(df['Accident_History'].mode()[0], inplace=True)
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['weather'] = le.fit_transform(df['weather'])
df['time_of_day'] = le.fit_transform(df['time_of_day'])
df['traffic_density'] = le.fit_transform(df['traffic_density'])
df['road_type'] = le.fit_transform(df['road_type'])
df['road_condition'] = le.fit_transform(df['road_condition'])
df['accident_history'] = le.fit_transform(df['accident_history'])
df['road_safe_to_travel'] = le.fit_transform(df['road_safe_to_travel'])
```

```python
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
import pickle

# assume your trained model variable name is: model
pickle.dump(model, open("model.pkl", "wb"))
```
